```python
import numpy as np
```

```python
entradas = np.array([0.5, 0.8, 0.2])
```

```python

pesos = np.array([0.3, -0.2, 0.7])
```

```python
vies = 0.1
```

```python
def perceptor(entradas, pesos, vies):
    soma = np.dot(entradas, pesos) + vies
    return soma
```

```python
resultado = perceptor(entradas, pesos, vies)
print(resultado)
```

```python
def relu(x):
   return np.maximum(0,x)
```

```python
saida_ativacao = relu(resultado)
print(saida_ativacao)
```

```python
def ativacao(x):
  if x > 0:
    return 1
  else:
      return 0
```

```python
saida_ativacao = ativacao(resultado)
print(saida_ativacao)
```
